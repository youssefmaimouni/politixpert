```python
df["date"] = pd.to_datetime(
    df["date"],
    errors="coerce",   # valeurs invalides โ NaT
    utc=False
)

# Garder uniquement la date (sans heure)
df["date"] = df["date"].dt.date

# -----------------------------
# Vรฉrifier le date range
# -----------------------------

date_min = df["date"].min()
date_max = df["date"].max()

print("๐ Date minimale :", date_min)
print("๐ Date maximale :", date_max)
```

```python
import re

# -----------------------------
# Fonctions de nettoyage arabe
# -----------------------------

def remove_diacritics(text):
    """Supprimer les diacritiques arabes"""
    arabic_diacritics = re.compile("""
        ู|ู|ู|ู|ู|ู|ู|ู|ู
    """, re.VERBOSE)
    return re.sub(arabic_diacritics, '', text)

def normalize_arabic(text):
    """Normalisation des lettres arabes"""
    text = re.sub("[ุฅุฃุขุง]", "ุง", text)
    text = re.sub("ู", "ู", text)
    text = re.sub("ุค", "ู", text)
    text = re.sub("ุฆ", "ู", text)
    text = re.sub("ุฉ", "ู", text)  # optionnel
    return text

def remove_urls(text):
    """Supprimer les liens"""
    return re.sub(r'http\S+|www\S+', '', text)

def remove_special_chars(text):
    """Supprimer les caractรจres non arabes"""
    text = re.sub(r"[^\u0600-\u06FF\s]", " ", text)
    return text

def normalize_spaces(text):
    """Normaliser les espaces"""
    return re.sub(r'\s+', ' ', text).strip()

def clean_arabic_text(text):
    if pd.isna(text):
        return ""
    text = str(text)
    text = remove_urls(text)
    text = remove_diacritics(text)
    text = normalize_arabic(text)
    text = remove_special_chars(text)
    text = normalize_spaces(text)
    return text

# Appliquer le nettoyage sur la colonne content
df["content"] = df["content"].apply(clean_arabic_text)
df["title"] = df["title"].apply(clean_arabic_text)
df["description"] = df["description"].apply(clean_arabic_text)


df
```

```python
import ast

df["categories"] = df["categories"].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

# Une catรฉgorie par ligne
df_exploded = df.explode("categories")
category_counts = (
    df_exploded["categories"]
    .value_counts()
    .reset_index()
)

category_counts.columns = ["category", "nb_articles"]

print(category_counts.head(20))
```

```python
import matplotlib.pyplot as plt
top_categories = category_counts.head(20)

plt.figure(figsize=(10, 6))
plt.barh(top_categories["category"], top_categories["nb_articles"])
plt.xlabel("Nombre d'articles")
plt.ylabel("Catรฉgorie")
plt.title("Top 20 catรฉgories par nombre d'articles")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
